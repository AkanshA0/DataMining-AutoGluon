## Load data

Load the train and test data from the `/content/CaliforniaHousingPrices` directory into pandas DataFrames.


```python
import pandas as pd

directory = '/content/CaliforniaHousingPrices/'
train_data = pd.read_csv(directory + 'train.csv')
test_data = pd.read_csv(directory + 'test.csv')
```

## Identify target variable

Identify the target variable for prediction in the California Housing Prices dataset.

## Initialize and train autogluon predictor

Initialize the `TabularPredictor` with the appropriate target variable and train it on the loaded training data.


```python
from autogluon.tabular import TabularPredictor

label = 'Sold Price'
save_path = directory + 'AutoGluonModels_housing/'

predictor = TabularPredictor(label=label, path=save_path, verbosity=3).fit(
    train_data, presets='best_quality', time_limit=600
)

results = predictor.fit_summary()
```

```python
display(results['leaderboard'])
```

Although we did not explicitly specify eval_metric='root_mean_squared_error' in the TabularPredictor, AutoGluon automatically detected that the problem is a regression task based on the continuous nature of the target variable ('Sold Price'). For regression problems, AutoGluon defaults to using RMSE as the primary evaluation metric to optimize model performance during training.Root Mean Squared Error (RMSE) is the appropriate evaluation metric for this task because we are performing **regression**, which involves predicting a continuous numerical value ('Sold Price'). RMSE measures the average magnitude of the errors between the predicted and actual values, providing a relevant measure of model performance for this type of problem. In contrast, metrics like ROC AUC are used for classification tasks, not regression.

```python
predictions = predictor.predict(test_data)
display(predictions.head(5))
```

## Generate Submission File

Create a submission file in the specified format with the predicted values.

```python
submission_df = pd.DataFrame({'Id': test_data['Id'], 'Sold Price': predictions})
submission_path = directory + 'submission.csv'
submission_df.to_csv(submission_path, index=False)

print(f"Submission file created successfully at: {submission_path}")
```
